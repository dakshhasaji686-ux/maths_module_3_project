# Part A – Theoretical Foundation

## 1. What is Inferential Statistics?
Inferential statistics is used to draw conclusions about a population using sample data. It helps in prediction and decision-making.

## 2. What is Hypothesis Testing and its Components?
Hypothesis testing is a method to check whether a claim is true or false using data. Its components are null hypothesis, alternative hypothesis, significance level, test statistic, and p-value.

## 3. Explain Confidence Interval and Critical Value.
A confidence interval is a range that likely contains the true population value. A critical value is used to decide whether to reject the null hypothesis.

## 4. Define p-value.
The p-value is the probability of getting the observed result if the null hypothesis is true. A small p-value indicates strong evidence against the null hypothesis.

## 5. Differentiate Type I and Type II Errors.
Type I error means rejecting a true null hypothesis. Type II error means accepting a false null hypothesis.

## 6. Brief descriptions of z-test, t-test, chi-square test, and ANOVA test.
- z-test: Compares means for large samples.
- t-test: Compares means for small samples.
- Chi-square test: Tests relationship between categorical variables.
- ANOVA: Compares means of three or more groups.

## 7. What is Covariance?
Covariance shows how two variables change together. Positive covariance means both move in the same direction.

## 8. What is Correlation?
Correlation measures the strength and direction of relationship between two variables. Its value ranges from -1 to +1.

```python
df=pd.read_csv("/content/healthcare_dataset_100_rows.csv")
df.head()
```

```python
from scipy import stats
from statsmodels.stats.weightstats import zconfint
from scipy.stats import ttest_ind, chi2_contingency, f_oneway

print("95% Confidence Interval")

numerical_cols = [
    'age',
    'weight',
    'bmi',
    'blood_pressure',
    'cholesterol_level',
    'glucose_level'
]

for col in numerical_cols:
    ci = zconfint(df[col])
    print(f"{col}: {ci}")
```

```python
# 3. Critical Value and p-value Example

alpha = 0.05
critical_value = stats.norm.ppf(1 - alpha/2)

print("\nCritical Value:", critical_value)
print("Significance Level:", alpha)

```

```python
# 4. t-test
# Compare BMI of Smokers vs Non-Smokers

smokers = df[df['smoking_status'] == 'Smoker']['bmi']
non_smokers = df[df['smoking_status'] == 'Non-Smoker']['bmi']

t_stat, p_value = ttest_ind(smokers, non_smokers)

print("\nT-Test Result")
print("T-Statistic:", t_stat)
print("P-Value:", p_value)

if p_value < 0.05:
    print("Reject H0")
else:
    print("Accept H0")
```

```python
# 5. Chi-Square Test
# Smoking Status vs Diabetes
contingency_table = pd.crosstab(df['smoking_status'], df['diabetes'])

chi2, p, dof, expected = chi2_contingency(contingency_table)

print("\nChi-Square Test")
print("Chi-Square Value:", chi2)
print("P-Value:", p)

if p < 0.05:
    print("Reject H0")
else:
    print("Accept H0")
```

```python
# 6. ANOVA Test
# BMI Across Age Groups

group1 = df[df['age_group'] == '18-25']['bmi']
group2 = df[df['age_group'] == '26-35']['bmi']
group3 = df[df['age_group'] == '36-45']['bmi']
group4 = df[df['age_group'] == '46-60']['bmi']
group5 = df[df['age_group'] == '60+']['bmi']

anova_result = f_oneway(group1, group2, group3, group4, group5)

print("\nANOVA Test")
print("F-Statistic:", anova_result.statistic)
print("P-Value:", anova_result.pvalue)

if anova_result.pvalue < 0.05:
    print("Reject H0")
else:
    print("Accept H0")
```

```python
# 7. Covariance and Correlation

covariance = np.cov(df['age'], df['bmi'])[0][1]
correlation = np.corrcoef(df['age'], df['bmi'])[0][1]

print("\nCovariance:", covariance)
print("Correlation:", correlation)

```
